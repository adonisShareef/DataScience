```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import squarify 
```

```python
df = pd.read_excel('data/US_General_Aviation.xlsx', index_col=0)  
df
```

```python
#bar chart
df['Total fatalities'].plot(kind='bar', title='Total fatalities', figsize=(20,10))
```

```python
#scatter plot
plt.scatter(df['Total accidentsa,b'], df['Total accidentsa,b, fatal'])
plt.title('Total Fatal Accidents by Total Accidents')
plt.xlabel('Total accidents')
plt.ylabel('Total fatal accidents')
plt.show()
```
